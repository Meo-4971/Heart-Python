```python
# H3: Hút thuốc có phải là nguyên nhân dẫn đến bệnh tim nhanh hơn so với uống rượu? - Smoking

data_smoke = data[data['Smoking'] == "Yes"] # Lọc ra những người có hút thuốc
sum_smoke = len((data_smoke).index) #Tổng số người hút thuốc
smoke_Heart = len((data_smoke[data_smoke['HeartDisease'] == "Yes"]).index) # Tổng số người hút thuốc bị bệnh tim
smoke_noHeart = len((data_smoke[data_smoke['HeartDisease'] == "No"]).index) # Tổng số người hút thuốc ko bị bệnh tim

# Tỷ lệ người hút thuốc bị [0] và ko bị[1] bệnh tim
ratio_smoke = (round((smoke_Heart*100)/sum_smoke, 2), round((smoke_noHeart*100)/sum_smoke), 2)


data_alcohol = data[data['AlcoholDrinking'] == "Yes"]
sum_alcohol = len((data_alcohol).index) #Tổng số người uống rượu
alcohol_Heart = len((data_alcohol[data_alcohol['HeartDisease'] == "Yes"]).index) #Tổng số người uống rượu bị bệnh tim
alcohol_noHeart = len((data_alcohol[data_alcohol['HeartDisease'] == "No"]).index) #Tổng số người uống rượu ko bị bệnh tim

# Tỷ lệ người nghiện rượu bị [0] và ko bị[1] bệnh tim
ratio_alcohol = (round((alcohol_Heart*100)/sum_alcohol, 2), round((alcohol_noHeart*100)/sum_alcohol, 2))

df_nghien1 = pd.DataFrame([[smoke_Heart, alcohol_Heart],
                        [smoke_noHeart, alcohol_noHeart]],
                        index = ['Bị bệnh tim', 'Không bị bệnh tim'], 
                        columns = ['Tổng số người hút thuốc lá', 'Tổng số người nghiện rượu'])
df_nghien2 = pd.DataFrame([[ratio_smoke[0], ratio_alcohol[0]],
                        [100-ratio_smoke[0], 100 - ratio_alcohol[0]]],
                        index = ['Bị bệnh tim', 'Không bị bệnh tim'], 
                        columns = ['Tỷ lệ người hút thuốc lá', 'Tỷ lệ người nghiện rượu'])
```

# Bước 3: Khoảng tin cậy
- Sử dụng khoảng tin cậy 95% để ước lượng khoảng trung bình độ tuổi, trung bình BMI, tỷ lệ giữa người hút thuốc lá và nghiện rượu tới bệnh tim.

```python
confidence = 0.95
t1 = 0
t2 = 0
t3 = 0
t4 = 0
t5 = 0
t6 = 0
t7 = 0
t8 = 0

def mean_confidence_interval(data):
    n = len(data)
    mean, SE = np.mean(data), scipy.stats.sem(data) # trung bình và sai số chuẩn
    sample_SD = stat.stdev(data) # độ lệch chuẩn mẫu
    #t_value
    t_value = stats.t.ppf(1 - (1 - confidence)/2, n-1)
    sampling_dist_mean_SD = sample_SD/math.sqrt(n)
    h = SE * scipy.stats.t.ppf((1 + confidence) / 2, n-1)
    # print((mean-h, mean+h))
    m1 = mean - h
    m2 = mean + h
    return t_value * sampling_dist_mean_SD, m1, m2

def draw_khoang_tin_cay(mean_lt, error_lt, labels):
    fig = plt.figure(figsize=(10,10))
    plt.errorbar(x=mean_lt, y=range(0,len(mean_lt)), xerr=error_lt, fmt='o', color='k')
    plt.yticks([0,1],labels=labels)
    plt.grid(linewidth=1)
    plt.show();

data_Heart_error, t1, t2 = mean_confidence_interval(data_Heart.Age)
data_NoHeart_error, t3, t4 = mean_confidence_interval(data_NoHeart.Age)
error_lt = [data_Heart_error, data_NoHeart_error]
mean_lt = [stat.mean(data_Heart.Age), stat.mean(data_NoHeart.Age)]
labels = ["Khoảng tin cậy\n Heart Disease","Khoảng tin cậy\n No Heart Disease"]

data_HeartBMI_error, t5, t6 = mean_confidence_interval(data_Heart.BMI)
data_NoHeartBMI_error, t7, t8 = mean_confidence_interval(data_NoHeart.BMI)
error_lt = [data_HeartBMI_error, data_NoHeartBMI_error]
mean_lt = [stat.mean(data_Heart.BMI), stat.mean(data_NoHeart.BMI)]
labels = ["Khoảng tin cậy\n Heart Disease","Khoảng tin cậy\n No Heart Disease"]
```

```python
print("\tƯớc lượng giá trị trung bình của độ tuổi")
print("Trung bình độ tuổi của người bị bệnh tim nằm trong khoảng: (", t1, ",", t2, ")")
print("Trung bình độ tuổi của người không bị bệnh tim nằm trong khoảng: (", t3, ",", t4, ")")
draw_khoang_tin_cay(mean_lt,error_lt,labels)
```

```python
print("\tƯớc lượng giá trị trung bình của chỉ số khối cơ thể")
print("Trung bình chỉ số khối của người bị bệnh tim nằm trong khoảng: (", t5, ",", t6, ")")
print("Trung bình chỉ số khối của người không bị bệnh tim nằm trong khoảng: (", t7, ",", t8, ")")
draw_khoang_tin_cay(mean_lt,error_lt,labels)
```

```python
z_value = norm.ppf(0.975)
#Tính khoảng tin cậy tỉ lệ cho smoking_heart
p_smoking_heart = ratio_smoke[0]/100
smoking_heart_SE = math.sqrt((p_smoking_heart*(1 - p_smoking_heart))/smoke_Heart)
smoking_heart_error = smoking_heart_SE*z_value

#Tính khoảng tin cậy tỉ lệ cho alcohol_heart
p_alcohol_heart = ratio_alcohol[0]/100
alcohol_heart_SE = math.sqrt((p_alcohol_heart*(1 - p_alcohol_heart))/alcohol_Heart)
alcohol_heart_error = alcohol_heart_SE*z_value

data_dict = {}
data_dict['category'] = ['smoke_Heart', 'alcohol_Heart']
data_dict['left'] = [p_smoking_heart - smoking_heart_error, p_alcohol_heart - alcohol_heart_error]
data_dict['right'] = [p_smoking_heart + smoking_heart_error, p_alcohol_heart + alcohol_heart_error]
dataset = pd.DataFrame(data_dict)
```

```python
for lower,upper,y in zip(dataset['left'],dataset['right'],range(len(dataset))):
    plt.plot((lower,upper),(y,y), 'o-', color='black')
plt.yticks(range(len(dataset)),list(dataset['category']))
plt.title('Khoảng tin cậy tỉ lệ', color = 'black', fontsize = 13)
plt.grid(linewidth=1)

print('Tỉ lệ người hút thuốc bị bệnh tim nằm trong khoảng: (', data_dict['left'][0], ', ', data_dict['right'][0], ')')
print('Tỉ lệ người nghiện rượu bị bệnh tim nằm trong khoảng: (', data_dict['left'][1], ', ', data_dict['right'][1], ')')
```

### Nhận xét: 
Dựa vào khoảng tin cậy 95% ta có thể thấy rằng:
- Trung bình của độ tuổi người bị bệnh tim (68.85, 69.54) lớn hơn so với người không bị bệnh tim (53.82, 54.13).
- Trung bình của chỉ số BMI người bị bệnh tim (28.92, 29.28) chênh lệch không nhiều so với người không bị bệnh tim (27.92, 28.03).
- Tỷ lệ người hút thuốc lá bị bệnh tim (0.1145, 0.1375) cao hơn hẳn so với tỷ lệ người nghiện rượu bị bệnh tim (0.0313,  0.0907).

```python
ageLon_Heart0 = data_ageLon[data_ageLon.HeartDisease == "Yes"] # Người lớn tuổi bị bệnh tim
ageNho_Heart0 = data_ageNho[data_ageNho.HeartDisease == "Yes"] # Người nhỏ tuổi bị bệnh tim

print("H0: Khả năng mắc bệnh tim của người cao tuổi và người trẻ tuổi là như nhau")
print("H1: Người lớn tuổi có khả năng mắc bệnh tim cao hơn người trẻ tuổi")
t_stat, p = stats.ttest_ind(ageLon_Heart0.Age, ageNho_Heart0.Age)
print(f"Giá trị p tính được là kiểm định hai phía: p = {p}")
print(f"Vì giả thuyết là kiểm định một phía nên giá trị p = p / 2 = {p} / 2 = {p/2}")
p = p / 2
print(f"Ta có p_value = {p} < 0.05 ==> Ta có thể bác bỏ H0 và chấp nhận H1")
```

```python
total_heart = smoke_Heart + alcohol_Heart
total_NoHeart = smoke_noHeart + alcohol_noHeart
total = total_heart + total_NoHeart

def tinhKyVong(total_r, total_c, total):
    return (total_r*total_c)/total

smoke_Heart_ex = tinhKyVong(sum_smoke, total_heart, total)
smoke_NoHeart_ex = tinhKyVong(sum_smoke, total_NoHeart, total)
alcohol_Heart_ex = tinhKyVong(sum_alcohol, total_heart, total)
alcohol_NoHeart_ex = tinhKyVong(sum_alcohol, total_NoHeart, total)

total_heart_ex = smoke_Heart_ex + alcohol_Heart_ex
total_NoHeart_ex = smoke_NoHeart_ex + alcohol_NoHeart_ex
total_ex = total_heart_ex + total_NoHeart_ex

def chi_squared(obs, exp):
    return ((obs - exp)**2)/exp

squared = chi_squared(smoke_Heart, smoke_Heart_ex) + chi_squared(smoke_noHeart, smoke_NoHeart_ex) + chi_squared(alcohol_Heart, alcohol_Heart_ex) + chi_squared(alcohol_noHeart, alcohol_NoHeart_ex)
p_value = 1 - stats.chi2.cdf(squared, 1)
```

```python
print("H0: Người hút thuốc lá có khả năng mắc bệnh tim ngang bằng so với người nghiện uống rượu.")
print("H3: Người hút thuốc lá có khả năng mắc bệnh tim cao hơn so với người nghiện uống rượu.")

if (p_value > 0.05):
    print(f"\nTa có p_value = {p_value} > 0.05 ==> Ta chấp nhận H0")
else:
    print(f"\nTa có p_value = {p_value} < 0.05 ==> Ta bác bỏ H0, chấp nhận H3")
```
